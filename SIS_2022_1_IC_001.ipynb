```python
def outliers(classe, atributo, valor):
  tmp = df[df['class']==classe]
  return tmp[tmp[atributo]<valor]
```

```python
import numpy as np
def remove_outlier(df):
  for coluna in df.columns[1:-1]:
    for especie in df['class'].unique():
      df_especie = df[df['class']==especie]
      df_colunas = df_especie[coluna]
      Q1 = df_colunas.quantile(0.25)
      Q3 = df_colunas.quantile(0.75)
      IQR = Q3-Q1
      print(f'Antes de excluir: {df.shape}')
      # obtendo os outliers => anomalias => valores fora do padrao
      anomalias = df_colunas[(df_colunas>(Q3+1.5*IQR)) | (df_colunas<(Q1-1.5*IQR))].index
      df.drop(index=anomalias, inplace = True)
  return df
```

```python
df = remove_outlier(df)
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
box_plot(df,'sepal_length')
plt.subplot(222)
box_plot(df,'sepal_width')
plt.subplot(223)
box_plot(df,'petal_length')
plt.subplot(224)
box_plot(df,'petal_width')
```
